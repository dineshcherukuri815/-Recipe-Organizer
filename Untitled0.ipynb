### Exploratory Data Analysis (EDA)

### Sales Trend Over Time

```python
plt.figure(figsize=(14, 6))

# Group by order date and sum sales
daily_sales = df.groupby('Order Date')['Sales'].sum().reset_index()

# Sort by date to ensure correct plotting
daily_sales.sort_values('Order Date', inplace=True)

# Plot daily sales
plt.plot(daily_sales['Order Date'], daily_sales['Sales'], label='Daily Sales', alpha=0.6)

# Add rolling average (7-day)
daily_sales['Rolling Avg'] = daily_sales['Sales'].rolling(window=7).mean()
plt.plot(daily_sales['Order Date'], daily_sales['Rolling Avg'], color='orange', label='7-Day Rolling Average')

# Titles and labels
plt.title("Daily Sales Over Time", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.tight_layout()
plt.show()
```

### Year-wise Total Sales

```python
# Ensure 'Order Date' is datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Extract year
df['Year'] = df['Order Date'].dt.year

# Aggregate sales by year
yearly_sales = df.groupby('Year')['Sales'].sum().reset_index().sort_values('Year')

# Plot
plt.figure(figsize=(12, 6), constrained_layout=True)
sns.barplot(data=yearly_sales, x='Year', y='Sales', color='skyblue')

# Add value labels on bars
for index, row in yearly_sales.iterrows():
    plt.text(index, row['Sales'] + 1000, f"{int(row['Sales'])}", ha='center', va='bottom', fontsize=10)

# Title and labels
plt.title("Year-wise Total Sales", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Total Sales", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

Regional Sales Analysis

```python
# Group sales by Region
region_sales = df.groupby('Region')['Sales'].sum().reset_index().sort_values(by='Sales', ascending=False)

# Plot
plt.figure(figsize=(10, 5))
sns.barplot(data=region_sales, x='Region', y='Sales', palette='Set2')

# Labels and title
plt.title("Total Sales by Region", fontsize=16)
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()
```

### Total Sales Category

```python
# Group sales by Category
category_sales = df.groupby('Category')['Sales'].sum().reset_index().sort_values(by='Sales', ascending=False)

# Plot
plt.figure(figsize=(8, 5))
bars = plt.bar(category_sales['Category'], category_sales['Sales'], color='lightcoral')
plt.title("Total Sales by Category", fontsize=14)
plt.xlabel("Category")
plt.ylabel("Sales")

# Add value labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f'{height:,.0f}', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

### Total Sales By Sub-Category

```python

# Group and sort total sales by sub-category
subcat_sales = df.groupby('Sub-Category')['Sales'].sum().reset_index()
subcat_sales = subcat_sales.sort_values(by='Sales', ascending=False)

# Calculate total sales for percentage
total_sales = subcat_sales['Sales'].sum()
subcat_sales['Percent'] = (subcat_sales['Sales'] / total_sales) * 100

# Plot
plt.figure(figsize=(14, 6))
ax = sns.barplot(data=subcat_sales, x='Sub-Category', y='Sales', palette='Set2')

# Add % labels on top of bars
for bar, pct in zip(ax.patches, subcat_sales['Percent']):
    height = bar.get_height()
    ax.annotate(f'{pct:.1f}%',  # 1 decimal place
                (bar.get_x() + bar.get_width() / 2, height),
                ha='center', va='bottom', fontsize=10)

# Titles and labels
plt.title("Total Sales by Sub-Category", fontsize=16)
plt.xlabel("Sub-Category")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
